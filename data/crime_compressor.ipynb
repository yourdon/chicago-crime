```python
import pandas as pd
import numpy as np
```

```python
# police budget from 2001 to 2010; from annual reports
police_budget = [940936247, 1014762072, 1051226834, 1073131416, 1066250557,
                 1217401602, 1247932269, 1217139440, 1234586462, 1242139543]
```

```python
# police personnel from 2001 to 2010; from annual reports
police_personnel = [16742, 16473, 16244, 15675, 15533,
                    15661, 15493, 15317, 14973, 13857]
```

```python
df = pd.read_csv("Crimes_-_2001_to_present.csv", usecols=[5,8, 17],
                 dtype = {"Year": pd.np.int64, "Primary.Type" : pd.np.object, "Arrest" : pd.np.bool_})
```

```python
crime_count = {}
crimes = ["HOMICIDE", "ASSAULT", "THEFT",
          "NARCOTICS", "BURGLARY", "BATTERY",
          "CRIM SEXUAL ASSAULT", "ARSON",
          "MOTOR VEHICLE THEFT", "PROSTITUTION",
          "CRIMINAL DAMAGE", "ARRESTS", "ALL"]
```

```python
for i in range(2001,2011):
    crime_count[i] = [0]*len(crimes)
```

```python
for i in df.iterrows():
    if i[1]["Year"] > 2010:
        continue
    for j in range(len(crimes)-1):
        if i[1]["Primary Type"] == crimes[j]:
            crime_count[i[1]["Year"]][j] += 1
    if i[1]["Arrest"] == True:
        crime_count[i[1]["Year"]][len(crimes)-2] += 1
    crime_count[i[1]["Year"]][len(crimes)-1] += 1
```

```python
crime_list = []
k = 0
for i in crime_count.items():
    crime_list.append([i[0]])
    for j in range(len(crimes)):
        crime_list[k].append(i[1][j])
    k += 1
```

```python
crime_df = pd.DataFrame(crime_list, columns=['Year'] + crimes)
crime_df = crime_df.dropna()
```

```python
crime_df['Police Personnel'] = pd.Series(np.array(police_personnel), index = crime_df.index)
crime_df['Police Budget'] = pd.Series(np.array(police_budget), index = crime_df.index)
```

```python
crime_df
```

```python
crime_df_dev = crime_df.copy(deep=True)
```

```python
for k in crime_df_dev.columns[1:]:
    column_mean = np.mean(crime_df[k])
    for m in range(len(crime_df)):
        crime_df_dev[k].iloc[m] = 100.0*(crime_df_dev[k].iloc[m] - column_mean) / column_mean
```

```python
crime_df_dev.to_csv(path_or_buf="crime_deviations.csv", index=False)
```
